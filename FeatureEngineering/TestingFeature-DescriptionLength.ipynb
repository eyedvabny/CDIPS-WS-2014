# Selecting relevant numerical and categorical features for training

**custom functions:** getFeature.py, testFeature.py

**Workflow:**

1. With the usage of the modified sample script, randomly select a defined number of row from the training dataset
2. Retrieve only fields that is necessary for the feature engineering. + response variable
3. Process fields as necessary
4. Submit feature/response pairs to test
5. analyse result

```python
# Loading libraries:
import csv
import re
import scipy.sparse as sp
import numpy as np
import os
import random as rnd 
import getFeature as gf
import featureTest as ft
%matplotlib inline
import matplotlib.pyplot as plt
import datetime

# Setting up variables:
fields = ["itemid","category","subcategory","title","description","attrs",
          "price","is_proved","is_blocked", "phones_cnt", "emails_cnt", "urls_cnt", "close_hours"]

fileName = "avito_train.tsv"
itemsLimit = 100000
fieldOfInterest = ["is_blocked", "description", "price", "close_hours", "phones_cnt", "emails_cnt"]

# Calling getItems (modified version of the sample script)
response = []         # Field name: is_blocked
descriptionLength =[] # Field name: description
price = []            # Field name: price
closeHours = []       # Field name: close_hours
emails = []           # Field name: emails_cnt
phones = []           # Field name: phones_cnt
```

**Reading data:**

```python
for processedCnt, item in gf.getItems(fileName, itemsLimit, fieldOfInterest):
    response.append(int(item["is_blocked"]))
    descriptionLength.append(len(item["description"]))
    price.append(item["price"])
    closeHours.append(item["close_hours"])
    emails.append(item["emails_cnt"])
    phones.append(item["phones_cnt"])
```

### Testing if length of the descriptions correlates with the response variable

Before we can check if there is correlation between the two variables, 
I want to see how the distribution of the length of the descriptions look like.
How the feature looks like:

```python
print "Count of items: "+str(len(descriptionLength))+"\n"\
"Mean length of descriptions: "+str(np.round(np.mean(descriptionLength), 1))+" Words"+\
"\nMedian length of descriptions: "+str(np.round(np.median(descriptionLength), 1))+" Words"
```

Take a look at the distribution of the length of the descriptions.

```python
plt.hist(descriptionLength, bins=200)
plt.show()
```

Then let's separate the length values by the response variable ("is_blocked" == 1 or 0) then create the distribution:

```python
ft.NumericalFeatutreTest(descriptionLength, response, featureName = "Length of description", writeFile=True)
```

It seems that the distribution of the length is really similar regardless of the value of 
the "is_blocked" value. So the lengthe of the description is not a good feature.

### Testing if close hours correlates with the response variable

Now the question if the *"close_hours"* variable can be a good estimator for the response variable

```python
closeHours = [float(i) for i in closeHours] # treating values as floats instead of string
ft.NumericalFeatutreTest(closeHours, response, featureName = "CloseHours", writeFile=True)
```

There are correlation only in the extremes.

### Testing if price correlates with the response variable

```python
price = [float(i) for i in price] # treating values as floats instead of string
plt.hist(price, bins=np.logspace(0.1, 9, 100))
plt.gca().set_xscale("log")
plt.show()
```

The price range is insane: covers 9 order of magnitude. I will pool price values into log bins to see if there are any price ranges that we can use as predictors.
