### A code to visualize the predictive power of words or word combinations

Initializing variables and loading modules:

```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import random
import numpy as np
import processFeatures as pF
from sklearn.externals import joblib


%matplotlib inline
```

```python
# Reading dataset (sparse matrices saved by the sample script)
trainFeatures, trainTargets, trainItemIds, testFeatures, testItemIds = joblib.load("train_data.pkl")
```

```python
 def freqPlot(trainFeatures, trainTargets, getPoints=1000):
    '''
    this function processes two outputfeatures of the asana sample script:
    
    Input: trainFeatures (np.compressed sparse matrix), trainTargets (np.array)
    
    The function calculates the frequency of ads in which the ad was blocked: 
        1 = all ads, where this word occured was bocked, 
        0 = no blocked ads with that word
    '''
    
    # looping through the matrix:
    y = []
    x = []

    # Get the x dimension of the matrix (number of features):
    matrix_xDim = trainFeatures[1,:].todense()
    featureNumber = len(np.array(matrix_xDim).reshape(-1,))
    # print(featureNumber)
    if getPoints == "ALL":
        samples = range(featureNumber)
    else:
        samples = np.random.randint(0,featureNumber, size=getPoints)

    for i in samples:

        # select a column from the matrix:
        feature = trainFeatures[:,i].todense()
        feature2 = np.array(feature).reshape(-1,)

        # creating a 
        df = pd.DataFrame({"feature" : feature2, 
                           "response" : trainTargets})
        # Sum
        sumFeatures = np.sum(df.feature)

        # number, where is_blocked = 1
        blockedNumber = np.sum(df["feature"][df.response == 1])

        # freq, where is_blocked = 1
        blockedFreq = blockedNumber / sumFeatures

        # Adding coordinates:
        y.append(blockedFreq)
        x.append(sumFeatures)

    return x,y

```

```python
x,y = freqPlot(trainFeatures, trainTargets, getPoints="ALL")

fig = plt.figure()
ax = plt.subplot(111)

# Adding constant 1, where there is no response.
plt.scatter(x,y, s=2)
ax.set_ylabel('Frequency blocked', size=15)
ax.set_xlabel('Times it was used', size=15)
ax.set_xscale('log')
```
