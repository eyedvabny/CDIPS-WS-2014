### A list of functions that we use to characterize the correlations between features and the response variable

So this script is for testing simple categorical variables, that has only a limited values.

```python
def categoricalFeatureTest (feature, response, featureName = "Feature", writeFile=True):
    """
    feature: categorical variable, array, processed as scalar
    response: response variable, binary data: 0 or 1 is accepted in this version
    freatureName: name of the feature, optional, for documentation purpose
    writeFile: boolean, if True, the generated plot is saved as a png 
    """    
    
    # Testing input arrays if they are valid, at this point of the development, 
    # I assume, it is correct

    averageResponse = float(sum(response)) / len(response)
    
    # generating pandas dataframe:
    df = pd.DataFrame({featureName:feature, 
                       "response":response})
    
    # Number of features:
    featuresCount = len(set(feature))
    
    # Looping through all features:
    featureRatio = {}
    featureCount = {}
    for f in set(feature):
        responseForF = df[df[featureName] == f]["response"]
        featureRatio[f] = float(sum(responseForF)) / len(responseForF)
        featureCount[f] = len(responseForF)
    
    # PLotting response rates of each feature values:
    now = datetime.datetime.now()
    fig = plt.figure()
    ax = plt.subplot(111)
    plt.bar(range(featuresCount),featureRatio.values(), width=0.3, align="center")
    
    # setting x axis labels with the feature counts:
    xtickLabels = []
    for key in featureRatio.keys():
        xtickLabels.append(str(key)+"\n"+str(featureCount[key]))
    plt.xticks(range(featuresCount), xtickLabels)
    plt.tight_layout(pad=3.8)
    ax.plot([-0.5,featuresCount-0.5],[averageResponse,averageResponse],"r-", ms=3)
    ax.set_ylabel('Response rate', size=15)
    ax.set_title('Testing feature: '+featureName+"\n"+now.strftime("%m/%d %H:%M:%S"), size=15)
    
    # If desired, the plot is saved:    
    if writeFile:
        filename = "TestedFeature-"+featureName+"_"+now.strftime("%Y-%m-%d_%H%M%S")+".png"
        fig.savefig(filename,  )
        
    # Printing report:
    print "Number of items: "+str(len(feature))    
    print "Average response rate: "+str(averageResponse)    
    print "The tested feature has: "+str(featuresCount)+" values\nList of values:"
    for key in featureRatio.keys():
        print str(key)+" n="+str(featureCount[key])+" response rate="+str(featureRatio[key])
```

The following function analyzes the predictive power of a numerical feature. The feature binned into 200 bins, then separated by the response variable. The frequencies are divided, so we will see if any outcomes enriched in any bin.

```python
def NumericalFeatutreTest(feature, response, featureName = "Feature", responseName = "Response", writeFile=False):
    
    # In the future, a more intensive errorhandling will be implemened to make
    # sure the submitted arrays are proper
    
    # Dataframe is assambled from the input
    df = pd.DataFrame({featureName : feature,
                       responseName : response})
                       
    # getting extremes and do binning: 
    bins = 200
    minimum = min(feature)
    maximum = max(feature)

    response_1 = np.histogram(df[df[responseName] == 1][featureName], bins=bins, range=(minimum, maximum), density=True)
    response_0 = np.histogram(df[df[responseName] == 0][featureName], bins=bins, range=(minimum, maximum), density=True)
    
    relative_response = response_1[0]/response_0[0]
    
    # Plotting response rates of each feature values:
    now = datetime.datetime.now()
    fig = plt.figure()
    ax = plt.subplot(111)
    plt.tight_layout(pad=3.8)
    label=responseName+"= 1 / "+responseName+"= 0"
    
    # Adding constant 1, where there is no response.
    plt.plot(response_1[1][:-1],relative_response, "r-", label=label)
    plt.plot([response_1[1][0],response_1[1][-1]],[1,1])

    #plt.plot(response_0[0],response_0[1][:-1], "b-", label=responseName+"= 1, n="+str(len(df[df[responseName] == 1][featureName])) )
    
    ax.set_ylabel('Relative density', size=15)
    ax.set_xlabel('Feature value', size=15)
    ax.set_title('Testing numeric feature: '+featureName+"\n"+now.strftime("%m/%d %H:%M:%S"), size=15)
    plt.legend(loc='upper right')
    
    # If desired, the plot is saved:    
    if writeFile:
        filename = "TestedFeature-"+featureName+"_"+now.strftime("%Y-%m-%d_%H%M%S")+".png"
        fig.savefig(filename,  )
```

### Testing if categories and subcategories correlates with the response variable

In this case the process is a bit more complicated... the two colums are completely depend on each other, and before the test I want to concatenate the fields.

```python
PooledCategory = []

kindCategory = set(df.category) # alphabetically sorted! 
lenCategory = range(len(kindCategory))
kindSubcategory = set(df.subcategory)
lenSubcategory = range(len(kindSubcategory))

# We might want to dump out the dictionary.... and the 

# Creaging dictionaries:
CategoryDict = dict(zip(kindCategory,lenCategory))
SubcategoryDict = dict(zip(kindSubcategory,lenSubcategory))

def f(x):
    cat = x[0]
    subcat = x[1]
    return  str(CategoryDict[cat])+"_"+str(SubcategoryDict[subcat])

PooledCategory = df[["category","subcategory"]].apply(f, axis=1)

categoricalFeatureTest(PooledCategory, df.is_blocked, featureName="categories", writeFile=False)
```
